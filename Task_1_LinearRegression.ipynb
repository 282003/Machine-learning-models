```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing data 
data = pd.read_csv('task_1.csv')
data.head()
```

```python
df = data.drop('Scores', axis='columns')
```

```python
plt.scatter(data.Hours, data.Scores)
plt.title("Hours vs Scores")
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Scores')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df,data.Scores, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
#plot trendline 
line = model.coef_*df + model.intercept_

plt.scatter(df, data.Scores)
plt.plot(df, line)
```

```python
y_predict = model.predict(X_test)
model.score(X_test,y_test)
```

```python
comp_data = pd.DataFrame({'Actual':y_test,'Predicted':y_predict})
comp_data.head()
```

```python
#find the accuracy of model
from sklearn import metrics
print('Mean absolute error:', metrics.mean_absolute_error(y_test,y_predict))
```

```python
hours = [9.25]
hr = pd.DataFrame(hours)
hours_predict = model.predict(hr)
print('No. of hours studied:',hours[0])
print('Predicted percentage of score is:', hours_predict[0])
```
